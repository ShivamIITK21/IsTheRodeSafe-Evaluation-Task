```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Adult.csv', names=["Age", "Workclass", "fnlwgt", "education", "education-num", "martial-status", "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss", "hours", "country", "income"])
data.head(10)
```

```python
#data has missing values as ?
data = data.replace('?', np.NaN)
data = data.dropna()
print(data.shape)
x = data.iloc[:, 0:14]
y = data.iloc[:, 14]
x.head(10)
```

```python
categorical_cols = []
for col in x.columns:
    if x[col].dtypes == 'object':
        categorical_cols.append(col);

x = pd.get_dummies(x, columns=categorical_cols)
x = x.values
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

pipelines = {
    'rf':make_pipeline(StandardScaler(), RandomForestClassifier()),
    'lr':make_pipeline(StandardScaler(), LogisticRegression()),
    'kn':make_pipeline(StandardScaler(), KNeighborsClassifier()),
    'dt':make_pipeline(StandardScaler(), DecisionTreeClassifier())
}

fit_models = {}

for algo, pipeline in pipelines.items():
    model = pipeline.fit(x_train, y_train)
    fit_models[algo] = model
```

```python
from sklearn.metrics import accuracy_score
for algo, model in fit_models.items():
    y_pred = model.predict(x_test)
    print(algo, accuracy_score(y_test, y_pred))
```
